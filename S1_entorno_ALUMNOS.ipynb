## Ejercicio 10 — Tu primer gráfico 📊

````python
df["age"].hist(bins=20)
plt.title("Edades de los pasajeros del Titanic")
plt.xlabel("Edad")
plt.ylabel("Nº de pasajeros")
plt.show()
````
